```python
strikes_by_year = strikes.groupby('Incident Year')['Incident Year'].count().reset_index(name='total_strikes')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(strikes_by_year['Incident Year'], strikes_by_year['total_strikes'], color='blue', linewidth=1)
plt.xlabel('Year')
plt.ylabel('Number of Strikes')
plt.title('Wildlife Strikes by Year')
plt.grid(False)
plt.xticks([1990, 1995, 2000, 2005, 2010, 2015])
plt.yticks([2000, 6000, 10000, 14000])
plt.show()
```

```python
strikes_by_month = strikes.groupby('Incident Month')['Incident Month'].count().reset_index(name='total_strikes')
```

```python
plt.figure(figsize=(10, 6))
plt.plot(strikes_by_month['Incident Month'], strikes_by_month['total_strikes'], color='lightblue', linewidth=1, marker='o', markersize=6, markerfacecolor='blue')
plt.xlabel('Month')
plt.ylabel('Number of Strikes')
plt.title('Wildlife Strikes by Month')
plt.grid(False)
plt.xticks(range(1, 13))
plt.show()
```

```python
species = df["Species Name"]
species_count=species.value_counts()
print(species_count)
```

```python
species_count=species_count[species_count>4000]
print(species_count)
```

```python
top_species = ["UNKNOWN MEDIUM BIRD","UNKNOWN SMALL BIRD","MOURNING DOVE", "GULL","UNKNOWN BIRD","KILLDEER", "AMERICAN KESTREL","BARN SWALLOW"]
top_species = species[species.isin(top_species)]
print(top_species.value_counts())
```

```python
sns.countplot(top_species)
plt.title("Top Species That Impact with Aircraft")
plt.xticks(rotation='vertical')
```

```python
top_known_species = ["MOURNING DOVE", "GULL","KILLDEER", "AMERICAN KESTREL","BARN SWALLOW"]
top_known_species = species[species.isin(top_known_species)]
print(top_known_species.value_counts())
```
